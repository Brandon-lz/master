```
import tensorflow as tf
```

sigmoid 激活函数    输入过大或过小的时候输出接近于0或1，此时容易发生梯度消失的情况，使得优化变得非常缓慢

image.png

```
a = tf.linspace(-10.,10.,10)
```

```
with tf.GradientTape() as tape:
  tape.watch(a)
  y = tf.sigmoid(a)
grads = tape.gradient(y,[a])
grads
```

tanh 函数，可以通过sigmoid函数变化而来
image.png

注意：tanh和sigmoid的导数都可以通过函数本身得到，而不用求导

```
a = tf.linspace(-5,5,10)
tf.tanh(a)
```

ReLU函数： Rectified Linear Unit 非常简单的函数，但可以有效减少梯度爆炸或者梯度消失的情况
image.png

```
a = tf.linspace(-1,1,10)
tf.nn.relu(a)
```

```
#leaky_relu函数并不会把小于零的数置零，而是保持输入的符号，但会降低其斜率
b = tf.nn.leaky_relu(a)
b
```

```
import matplotlib.pyplot as plt
```

```
plt.plot(a,b)
plt.show()
```

```python
import tensorflow as tf
```

```python
#整型
tf.constant(1)
```

```python
#浮点型
tf.constant(1.)
```

```python
#类型不对会报错
tf.constant(2.2,dtype=tf.int32)
```

```python
tf.constant(1.2,dtype=tf.double)
```

```python
#布尔型
tf.constant([True,False])
```

```python
#1或0不能转化为bool型
tf.constant(1,dtype=tf.bool)
```

```python
tf.constant('hello world')
```

```python
#选择在cpu上操作
with tf.device('cpu'):
    a = tf.constant([1])

#选择再GPU上操作
with tf.device('gpu'):
    b = tf.range(4)
```

```python
a.device
```

```python
b.device
```

```python
#cpu转换到gpu上
aa = a.gpu()
```

```python
#gpu转换到cpu上
bb = b.cpu()
```

```python
b.ndim
```

```python
#判断是否为tensor数据
tf.is_tensor(a)
```

```python
import numpy as np
b = np.array([1.,2.,3.])
tf.is_tensor(b)
```

```python
#普通数据类型转换为tensor
bb = tf.convert_to_tensor(b)
tf.is_tensor(bb)
```

```python
bb
```

```python
#不同数据类型互相转换
tf.cast(bb,dtype=tf.int32)
```

```python
tf.cast(bb,dtype=tf.float32)
```

```python
tf.cast(bb,dtype=tf.double)
```

### 随机打散，重新排序

```python
#将有序数据打散为随机数据
idx = tf.range(10)
idx = tf.random.shuffle(idx)
idx
```

```python
a = tf.random.normal([10,784])
a
```

```python
#将数据a按照idx的顺序进行重新排列
a = tf.gather(a,idx)
a
```

```python
b = tf.random.uniform([10],maxval=10,dtype=tf.int32)
b
```

```python
#将数据b按照idx进行重新排列
b = tf.gather(b,idx)
b
```

### 卷积层的作用

```python
#假设有一个图片数据
x = tf.random.normal((4,32,32,3))        #3通道RGB，32宽，32高，一共4张图片
```

```python
net = layers.Conv2D(16,kernel_size=3)      #卷积，将数据维度拓宽
```

```python
net(x)          #(4, 30, 30, 16)
```

### 数据维度的意义

#### 三维数据
[6,5,5]   :    5个编码值，5个单词，6个句子 

#### 四维数据
[4,32,32,3] :   3个RGB通道，32个单位长度的宽度，32个高度，4张照片

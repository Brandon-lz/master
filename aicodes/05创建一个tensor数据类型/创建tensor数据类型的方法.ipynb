## 直接通过tensorflow创建

```python

tf.zeros([])    #创建一个标量
```

```python
tf.zeros([1])     #创建一个1维数据
```

```python
#注意，这里的[2,3]不是数据，而是要创建的数据的shape
tf.zeros([2,3])
```

```python
a = tf.ones([2,3,3])
a
```

```python
#创建一个shape相同，数据全0的数据
tf.zeros_like(a)
```

### 填充任意元素

```python
tf.fill([2,2],10)
```

```python
tf.fill([3,5],2.1)
```

### 创建正态分布数据

```python
tf.random.normal([2,2],mean=1,stddev=1)    #均值为1方差1
```

```python
tf.random.normal([2,2])      #标准正态分布
```

```python
tf.random.truncated_normal([2,2],mean=0,stddev=1)       #截断的正态分布
```

### 均匀分布

```python
tf.random.uniform([2,2],minval=0,maxval=1)
```

```python
tf.random.uniform([3,3],minval=-10,maxval=10)
```

```python
tf.random.uniform([3,3],minval=-10,maxval=10,dtype=tf.int32)    #指定数据类型
```

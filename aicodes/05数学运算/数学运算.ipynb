###  实现x@w+b运算

```python
x = tf.ones([4,2])
w = tf.ones([2,1])
b = tf.constant(0.1)
```

```python
x@w+b
```

```python
out = x@w+b
out = tf.nn.relu(out)     #加入非线性因子，relu函数，将负数去掉
out
```

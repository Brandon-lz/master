## Selective Indexing   更加灵活的采样，实现无规律index采样

### tf.gather:对某一个维度进行更加灵活采样，而其他维度保持不变

```python
tf.gather(a,axis=0,indices=[1,2]).shape       #对第一个维度，取1、2通道
```

```python
a[1:3].shape      #与上面等同
```

```python
tf.gather(a,axis=0,indices=[2,0,1])      #对第一个维度进行重新排序
```

```python
tf.gather(a,axis=2,indices=[2,3,7,9,0]).shape
```

### tf.gather_nd:对多个维度进行多通道采样

```python
a.shape
```

```python
tf.gather_nd(a,[0]).shape    #取第一个维度的第一个通道
```

```python
tf.gather_nd(a,[0,1]).shape     #取第一个维度的第一个通道，第二个维度的第二个通道
```

```python
tf.gather_nd(a,[0,1,12,15,2]).shape    #取1维度的1通道，二维度的2通道，三维度的13通道，四维度的16通道，最后一个维度的第三个通道，是一个标量
```

```python
tf.gather_nd(a,[[0,1,12,15,2]]).shape     #这里也是取出一个值，但是是向量
```

```python
a.shape
```

```python
tf.gather_nd(a,[[0,0],[1,1]]).shape    #第一维度取第一个通道，第二维度取第一个通道；第一维度取第二个通道，第二维度取第二个通道，然后拼接起来
```

```python
tf.gather_nd(a,[[0,0],[1,1],[2,3]]).shape
```

```python
tf.gather_nd(a,[[0,0,0],[1,1,2],[2,3,10]]).shape
```

### tf.boolean_mask ：类似tf.gather

```python
a.shape
```

```python
tf.boolean_mask(a,mask=[True,True,False]).shape     #对第一个维度的三个通道进行选择
```

```python
tf.boolean_mask(a,mask=[True,True,True,False],axis=1).shape     #对第二个维度的四个通道进行选取
```

```python
#对第一个维度的三个通道分别进行采样，每个通道下有四个第二维度的数据，所以总共是12个通道
tf.boolean_mask(a,mask=[[False,True,True,False],[True,False,True,False],[True,True,False,False]])     #一共取了六个通道
```

```python
a = tf.random.normal([4,28,28,3])
a.shape
```

#### 按一定间隔进行采样：


```python
a[:,:,:,::2]     #取0、2通道，隔一个采样一次
```

```python
a[0,...,2].shape
```

```python
#等同于
a[0,:,:,:,2].shape
```

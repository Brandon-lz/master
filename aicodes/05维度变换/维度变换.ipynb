### reshape

```python
tf.reshape(a,[4,784,3]).shape
```

```python
tf.reshape(a,[4,-1,3]).shape    #-1代表包含几个维度的所有通道数，但是只能使用一次-1
```

```python
tf.reshape(a,[4,784*3]).shape
```

```python
aa = tf.reshape(a,[4,-1])        #将数据维度打平
aa.shape
```

```python
#也可以用reshape将维度恢复
tf.reshape(aa,[4,28,28,3]).shape
```

### 维度变换

#### 增加维度

```python
a = tf.random.normal([4,35,8])
```

```python
tf.expand_dims(a,axis=0).shape         #在第一个维度前面增加一个维度
```

```python
tf.expand_dims(a,axis=3).shape      #从第一个维度的前面开始，从零往后数，第三个空上加一个维度
```

```python
tf.expand_dims(a,axis=-1).shape      #在最后一个维度后面加一个维度
```

```python
tf.expand_dims(a,axis=-2).shape
```

#### 减少维度

## 注意：对于shape=1的维度，可以直接省略

```python
#自动去掉所有通道为1的维度
tf.squeeze(tf.zeros([1,2,1,1,3])).shape   
```

```python
a = tf.zeros([1,2,1,3])
```

```python
tf.squeeze(a,axis=0).shape     #将第一个维度去掉
```

```python
tf.squeeze(a,axis=2).shape    #将第三个维度去掉
```

```python
tf.squeeze(a,axis=-2).shape
```

```python
tf.squeeze(a,axis=-1).shape      #如果删除通道数不是1的维度，则会报错
```

## Broadcasting    维度扩张

### 隐式Broadcasting，自动完成

```python

x = tf.random.normal([4,32,32,3])
```

```python
(x + tf.random.normal([3])).shape
```

```python
(x + tf.random.normal([32,32,1])).shape
```

```python
(x + tf.random.normal([4,1,1,1])).shape
```

```python
#无法隐式Broadcasting的情况
(x + tf.random.normal([1,4,1,1])).shape
```

### 显式Broadcasting

```python
x.shape
```

```python
b = tf.broadcast_to(tf.random.normal([4,1,1,1]),x.shape)
b.shape
```

### 利用tile实现Broadcasting

```python
a = tf.ones([3,4])
a.shape
```

```python
#用 broadcast_to实现
tf.broadcast_to(a,[2,3,4]).shape
```

```python
#用tile实现
a2 = tf.expand_dims(a,axis=0)
a2.shape
```

```python
tf.tile(a2,[2,1,1]).shape
```

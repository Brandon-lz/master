```python
import tensorflow as tf
```

## 合并操作

```python
a = tf.ones([4,35,8])
b = tf.fill([2,35,8],2.)
c = tf.concat([a,b],axis=0)       #axis=0代表将第一个维度进行合并，合并顺序前面a后面b
c
```

```python
a = tf.ones([4,32,8])
b = tf.fill([4,3,8],2.)
tf.concat([a,b],axis=1)
```

## 堆叠操作，堆出新维度

```python
a = tf.ones([4,35,8])
b = tf.fill([4,35,8],2.)
#注意：堆叠操作要求所有维度必须完全一致
```

```python
tf.stack([a,b],axis=0)       #axis=0代表堆叠的维度次序，这里是将所有的数据堆在新维度之后
```

```python
tf.stack([a,b],axis=3)        #将所有说堆在新维度之前，也就是最小维度为两通道数据
```

### 分解操作unstack

```python
a = tf.ones([4,35,8])
b = tf.ones([4,35,8])
```

```python
c = tf.stack([a,b])    #默认堆出来的新维度是最大维度
c.shape
```

```python
#分解操作
aa,bb = tf.unstack(c,axis=0)
aa.shape,bb.shape
```

```python
#在其他轴上进行分解
res = tf.unstack(c,axis=3)          #分解第三个维度
res[0].shape,res[5].shape, len(res)
```

### split，灵活性更强的unstack操作

```python
res = tf.split(c,axis=3,num_or_size_splits=2)       #可以设定分解的单位长度
len(res)
```

```python
res = tf.split(c,axis=3,num_or_size_splits=[2,2,4])       #甚至可以按照自定义长度进行分割
res[0].shape,res[2].shape, len(res)
```

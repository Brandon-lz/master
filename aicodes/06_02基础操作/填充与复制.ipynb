```python
import tensorflow as tf
```

```python
#二维tensor的pad
a = tf.reshape(tf.range(9),[3,3])
a
```

```python
tf.pad(a,[[1,0],[0,0]])     #对行pad，上面加一行，下面不加，对列pad，左右不加
```

```python
tf.pad(a,[[1,1],[0,0]])     #对行，上下各加一行
```

```python
tf.pad(a,[[1,0],[0,1]])
```

```python
#高维度tensor的pad
a = tf.random.normal([4,28,28,3])            #理解成四张图片
b = tf.pad(a,[[0,0],[2,2],[2,2],[0,0]])         #对图片的高和宽进行pad，上下左右都加2行像素
b.shape
```

```python
import tensorflow as tf
```

```python
#二维tensor的pad
a = tf.reshape(tf.range(9),[3,3])
a
```

```python
tf.pad(a,[[1,0],[0,0]])     #对行pad，上面加一行，下面不加，对列pad，左右不加
```

```python
tf.pad(a,[[1,1],[0,0]])     #对行，上下各加一行
```

```python
tf.pad(a,[[1,0],[0,1]])
```

```python
#高维度tensor的pad
a = tf.random.normal([4,28,28,3])            #理解成四张图片
b = tf.pad(a,[[0,0],[2,2],[2,2],[0,0]])         #对图片的高和宽进行pad，上下左右都加2行像素
b.shape
```

### tile、expand_dims、broadcast_to

```python
a = tf.random.uniform([3,3],maxval=10,dtype=tf.int32)
a
```

```python
tf.tile(a,[1,2])           #维度倍增
```

```python
tf.tile(a,[2,2])
```

```python
aa = tf.expand_dims(a,axis=0)    #增加维度
aa
```

```python
tf.broadcast_to(aa,[2,3,3])
```

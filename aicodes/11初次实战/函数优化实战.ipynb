首先绘制函数

```
def himmelblau(x):
  return (x[0]**2 + x[1]-11)**2 + (x[0]+x[1]**2 - 7)**2
```

```
import numpy as np
x = np.arange(-6,6,0.1)
y = np.arange(-6,6,0.1)
X, Y = np.meshgrid(x,y)
X.shape,Y.shape
```

```
Z = himmelblau([X,Y])
```

```
import matplotlib.pyplot as plt

fig = plt.figure('himmelblau')
ax = fig.gca(projection='3d')
ax.plot_surface(X,Y,Z)
ax.view_init(60,-30)
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```

Gradient Descent梯度下降

```
#x = tf.constant([-4.,0.])    #先给一个初始点x=-4,y=0，对于该初始点，我们发现可以在40以内的迭代次数中发现极小值
x = tf.constant([4.,0.])     #对于这个初始点，200次的迭代都没办法找到一个极小值
```

```
for step in range(200):
  with tf.GradientTape() as tape:
    tape.watch([x])
    y = himmelblau(x)
  grads = tape.gradient(y,[x])[0]      #grads是一个list，对于这里，grads中只有一个元素
  x -= 0.01*grads
  
  if step%20==0 :
    print('step:',step,'x=',x.numpy(),'f(x)=',himmelblau(x).numpy())
```

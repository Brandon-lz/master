## regularization

下式左边是交叉熵，右边是关于模型参数的范数，J作为loss放入学习过程中进行迭代优化。

这样做可以利用loss梯度减小的原理，将模型参数也缩小（weight decay），从而使整个模型的表达能力降低，也就是将有些高阶项的系数降低，从而达到简化模型的效果，拟合出来的函数曲线也会变得更加平滑。

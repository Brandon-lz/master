```python

def fourSum(nums, target):
    rtlist=[]
    for i in range(len(nums)):
        for j in range(i+1,len(nums)):
            nums1=nums.copy()
            nums1.pop(i)
            nums1.pop(j-1)

            for k in range(len(nums1)):
                for l in range(k+1,len(nums1)):
                    if target-nums[i]-nums[j]-nums1[k]==nums1[l]:
                        alist = [nums[i], nums[j], nums1[k], nums1[l]]
                        for m in range(3):
                            for n in range(4-1-m):
                                if alist[n]>alist[n+1]:
                                    alist[n], alist[n+1] = alist[n+1], alist[n]
                        rtlist.append(alist)
    for a in range(4):                    
        for i in range(len(rtlist)-1):
            for j in range(len(rtlist)-1-i):
                if rtlist[j][a]>rtlist[j+1][a]:
                    rtlist[j], rtlist[j+1] = rtlist[j+1], rtlist[j]
    print(rtlist)

    dellist=[]
    for i in range(len(rtlist)-1):
        if rtlist[i][0]==rtlist[i+1][0] and rtlist[i][1]==rtlist[i+1][1] and rtlist[i][2]==rtlist[i+1][2]:
            dellist.append(i)
    for i in dellist[::-1]:
        rtlist.pop(i)
    return rtlist
```

```python
#动态规划
import pandas as pd
import numpy as np
class CellClass(object):
    def __init__(self,price:int,things:list):
        self.price=price
        self.things=things
    def __str__(self):
        return str(self.price)+'-'+str(self.things)

#根据条件修改此处即可

things = pd.DataFrame([[0.5,7],[0.5,6],[1,9],[2,9],[0.5,8]],columns=['weight','price'],index=['威斯敏斯特教堂','环球剧场','英国国家美术馆','大英博物馆','圣保罗大教堂'])      #能偷的物品以及物品属性
colums = np.arange(0,3,0.5)[1:5]

'''
things = pd.DataFrame([[1,1500],[4,3000],[3,2000]],columns=['weight','price'],index=['吉他','音箱','笔记本电脑'])      #能偷的物品以及物品属性
colums = np.array([1,2,3,4])            #背包的空间根据粒度进行分割
'''

package=pd.DataFrame({n:[CellClass(0,[]) for i in colums] for n in things.index},index=colums).transpose()

#先填充第一行
for cell in package.columns:
    if things.loc[package.index[0]]['weight']<=cell:
        package.iloc[0].loc[cell].price=things.loc[package.index[0]].price
        package.iloc[0].loc[cell].things.append(package.index[0])
        
#填充剩下的行
for i in range(1, len(package.index)):          #i代表行数字索引
    for c in package.columns:       #c代表列索引，既列本身
        remain_weight = c - things.loc[package.index[i]].weight

        if remain_weight > 0:
            remain_price = package.iloc[i - 1][remain_weight].price                             #如果东西只有一件，则为i-1，如果可以重复拿，则是i
            if package.iloc[i - 1].loc[c].price >= (things.loc[package.index[i]].price + remain_price):
                package.iloc[i].loc[c].price = package.iloc[i - 1].loc[c].price
                package.iloc[i].loc[c].things = package.iloc[i - 1].loc[c].things.copy()
            else:
                package.iloc[i].loc[c].price = things.loc[package.index[i]].price + remain_price
                package.iloc[i].loc[c].things = [package.index[i]] + package.iloc[i - 1][remain_weight].things      #如果东西只有一件，则为i-1，如果可以重复拿，则是i

        elif remain_weight == 0:

            if package.iloc[i - 1].loc[c].price >= (things.loc[package.index[i]].price):
                package.iloc[i].loc[c].price = package.iloc[i - 1].loc[c].price
                package.iloc[i].loc[c].things = package.iloc[i - 1].loc[c].things.copy()
            else:
                package.iloc[i].loc[c].price = things.loc[package.index[i]].price
                package.iloc[i].loc[c].things = [package.index[i]]
        else:
            package.iloc[i].loc[c].price = package.iloc[i - 1].loc[c].price
            package.iloc[i].loc[c].things = package.iloc[i - 1].loc[c].things.copy()

package
```

```python
from tensorflow.keras.layers.normalization import BatchNormalization
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers.core import Dense,Dropout,Activation
from tensorflow.keras.optimizers import SGD,Adam
import numpy as np

def fizzbuzz(start,end):
    x_train,y_train=[],[]
    for i in range(start,end+1):
        num = i
        tmp=[0]*10
        j=0
        while num :
            tmp[j] = num & 1
            num = num>>1
            j+=1        
        x_train.append(tmp)
        if i % 3 == 0 and i % 5 ==0:
            y_train.append([0,0,0,1])
        elif i % 3 == 0:
            y_train.append([0,1,0,0])
        elif i % 5 == 0:
            y_train.append([0,0,1,0])
        else :
            y_train.append([1,0,0,0])
    return np.array(x_train),np.array(y_train)

x_train,y_train = fizzbuzz(101,1000) #打标记函数
x_test,y_test = fizzbuzz(1,100)

model = Sequential()
model.add(Dense(input_dim=10,units=100))
model.add(Activation('relu'))
model.add(Dense(units=4))
model.add(Activation('softmax'))

model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])

model.fit(x_train,y_train,batch_size=20,epochs=100)

result = model.evaluate(x_test,y_test,batch_size=1000)

print('Acc：',result[1])
```
